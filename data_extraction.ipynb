```python
all_match_data = [] # List to hold all processed match dictionaries

for file in tqdm(filenames):
    file_path = os.path.join('/content/t20s', file) # Use os.path.join for better path handling
    try:
        with open(file_path, 'r') as f:
            data = safe_load(f)
            if isinstance(data, dict): # Ensure the loaded data is a dictionary
                data['match_id'] = file.replace('.yaml', '')
                all_match_data.append(data)
            else:
                print(f"Skipping file {file}: Expected dictionary but got {type(data)}")
    except Exception as e:
        print(f"Error processing file {file}: {e}")

# Normalize all collected data at once
final_df = pd.json_normalize(all_match_data)

display(final_df.head())
```

```python
count = 1
# Initialize an empty list to store DataFrames
all_delivery_dataframes = []

for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []

    # Ensure 'innings' key exists and is a list before proceeding
    if 'innings' in row and isinstance(row['innings'], list) and len(row['innings']) > 0:
        first_innings = row['innings'][0].get('1st innings')
        if first_innings and 'deliveries' in first_innings and isinstance(first_innings['deliveries'], list):
            for ball in first_innings['deliveries']:
                for key in ball.keys():
                    match_id.append(count)
                    batting_team.append(first_innings['team'])
                    teams.append(row['info.teams'])
                    ball_of_match.append(key)
                    batsman.append(ball[key]['batsman'])
                    bowler.append(ball[key]['bowler'])
                    runs.append(ball[key]['runs']['total'])
                    city.append(row['info.city'])
                    venue.append(row['info.venue'])
                    try:
                        player_of_dismissed.append(ball[key]['wicket']['player_out'])
                    except:
                        player_of_dismissed.append('0')

            # Create a DataFrame for the current match's deliveries
            loop_df = pd.DataFrame({
                    'match_id':match_id,
                    'teams':teams,
                    'batting_team':batting_team,
                    'ball':ball_of_match,
                    'batsman':batsman,
                    'bowler':bowler,
                    'runs':runs,
                    'player_dismissed':player_of_dismissed,
                    'city':city,
                    'venue':venue
                })
            all_delivery_dataframes.append(loop_df)
        else:
            print(f"Skipping match_id {count} due to missing or invalid '1st innings' or 'deliveries' data.")
    else:
        print(f"Skipping match_id {count} due to missing or invalid 'innings' data.")

# Concatenate all collected DataFrames outside the loop
delivery_df = pd.concat(all_delivery_dataframes, ignore_index=True)
display(delivery_df.head())
```
